```python
from bs4 import BeautifulSoup  #help to extract data from web pages
```

```python
import requests   # used to send https request}
```

```python
import numpy as np
import pandas as pd
import matplotlib._animation_data as anm
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url="https://en.wikipedia.org/wiki/Economy_of_India"  #url of a web page
request_responce=requests.get(url)  #getting request
soup=BeautifulSoup(request_responce.text,'html.parser')  #we want text data

my_table=soup.find('table',{'class':"wikitable"})   #we have to extract data of table tag which has class 'wikitable'

```

```python
table=pd.read_html(str(my_table))  #changed html data into pandas list
table
```

```python
df=pd.DataFrame(table[0])  #converted from list to data frame
```

```python
df
```

```python
ax1=sns.scatterplot(df['Year'],df['GDP growth(real)'],label='X')
sns.scatterplot(df['Year'],df['GDP per capita(in US$ PPP)'],label='Y',ax=ax1)
```
